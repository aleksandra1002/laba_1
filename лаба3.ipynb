```python
import requests
import time
import pandas as pd
import matplotlib.pyplot as plt

# Глобальные переменные для подсчета ошибок и успешных запросов
success_count = 0
error_count = 0

# Функция для получения данных в формате FASTA
def uniprot_data(uniprot_id):
    global success_count, error_count
    query_link = f"https://rest.uniprot.org/uniprotkb/search?query={uniprot_id}&format=fasta"
    try:
        response = requests.get(query_link)
        if response.status_code == 200:  # Проверка успешности запроса
            success_count += 1
            return response.text  # Возвращаем данные в формате FASTA
        else:
            response.raise_for_status()  # Выбрасываем исключение при ошибке
    except requests.HTTPError as e:
        print(f"Ошибка при запросе {uniprot_id}: {str(e)}")
        error_count += 1
        return None

# Чтение идентификаторов из файла
def read_uniprot(file_path):
    df = pd.read_excel(file_path)  # Чтение Excel-файла
    return df["Entry"].tolist()  # Возвращаем список идентификаторов

# Сохранение данных в файл
def save_fasta(data, file_path):
    with open(file_path, "w") as fasta_file:
        for entry in data:
            if entry:  # Игнорируем пустые записи
                fasta_file.write(entry + "\n")

# Основная функция
def main():
    global success_count, error_count
    times = []
    # Путь к файлу с идентификаторами
    input_file = "C:\\Users\\sasha\\OneDrive\\Рабочий стол\\infa\\uniprot.xlsx"
    # Путь к выходному файлу FASTA
    output_file = "uniprot_sequences.fasta"

    # Чтение идентификаторов
    uniprot_ids = read_uniprot(input_file)

    # Сбор данных
    fasta_data = []
    start_time = time.time()  # Начало замера времени

    for uniprot_id in uniprot_ids:
        print(f"Данные для {uniprot_id}...")
        data = uniprot_data(uniprot_id)
        if data:
            fasta_data.append(data)
        time.sleep(1)  # Задержка между запросами

        end_time = time.time()  # Конец замера времени
        print(f"Время выполнения: {end_time - start_time:.2f} секунд")
        times.append(end_time - start_time)

    # Сохранение данных
    save_fasta(fasta_data, output_file)
    print(f"Данные сохранены в {output_file}")

    # Построение графиков
    plt.figure(figsize=(12, 5))

    # График 1: Зависимость времени от количества идентификаторов
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(times) + 1), times, marker='o')
    plt.xlabel('Количество идентификаторов')
    plt.ylabel('Время (секунды)')
    plt.title('Зависимость времени от количества идентификаторов')
    plt.grid()

    # График 2: Статистика успешных и неудачных запросов
    plt.subplot(1, 2, 2)
    plt.bar(["Успешные", "Неудачные"], [success_count, error_count], color=["green", "red"])
    plt.title("Статистика запросов")
    plt.xlabel("Тип запроса")
    plt.ylabel("Количество")

    plt.tight_layout()
    plt.show()

# Запуск программы
if __name__ == "__main__":
    main()
```

```python
import aiohttp
import asyncio
import time
import pandas as pd
import matplotlib.pyplot as plt

# Функция для получения данных в формате FASTA
async def fetch_uniprot_data(session, uniprot_id):
    query_link = f"https://rest.uniprot.org/uniprotkb/search?query={uniprot_id}&format=fasta"
    start_time = time.time()  # Замер времени начала запроса
    try:
        async with session.get(query_link) as response:
            if response.status == 200:
                data = await response.text()
                end_time = time.time()  # Замер времени завершения запроса
                return data, end_time - start_time  # Возвращаем данные и время выполнения
            else:
                response.raise_for_status()
    except aiohttp.ClientError as e:
        end_time = time.time()  # Замер времени завершения запроса (в случае ошибки)
        print(f"Произошла ошибка при запросе {uniprot_id}: {str(e)}")
        return None, end_time - start_time  # Возвращаем None и время выполнения

# Чтение идентификаторов из файла Excel
def read_uniprot_ids(file_path):
    df = pd.read_excel(file_path)  # Чтение Excel-файла
    return df["Entry"].tolist()  # Возвращаем список идентификаторов

# Сохранение данных в файл FASTA
def save_fasta(data, file_path):
    with open(file_path, "w") as fasta_file:
        for entry in data:
            if entry:  # Игнорируем пустые записи
                fasta_file.write(entry + "\n")

# Основная функция для выполнения запросов
async def run_requests(uniprot_ids):
    start_time = time.time()  # Общее время начала выполнения всех запросов
    async with aiohttp.ClientSession() as session:
        tasks = [fetch_uniprot_data(session, uniprot_id) for uniprot_id in uniprot_ids]
        results = await asyncio.gather(*tasks)
    end_time = time.time()  # Общее время завершения всех запросов

    # Разделяем результаты и время выполнения
    fasta_data = [result[0] for result in results]  # Данные FASTA
    fetch_times = [result[1] for result in results]  # Время выполнения каждого запроса

    # Подсчет успешных и неудачных запросов
    success_count = sum(1 for result in results if result[0] is not None)
    failure_count = len(results) - success_count
    
    return len(uniprot_ids), success_count, failure_count, end_time - start_time, fetch_times

# Функция для сбора результатов и построения графиков
async def plot_results():
    # Путь к вашему Excel-файлу
    input_file = "C:\\Users\\sasha\\OneDrive\\Рабочий стол\\infa\\uniprot.xlsx"

    
    # Чтение идентификаторов из файла
    uniprot_ids = read_uniprot_ids(input_file)
    
    # Выполнение запросов
    count, success, failure, elapsed_time, fetch_times = await run_requests(uniprot_ids)
    
    # Вывод статистики
    print(f"Всего идентификаторов: {count}")
    print(f"Успешных запросов: {success}")
    print(f"Неудачных запросов: {failure}")
    print(f"Общее время выполнения: {elapsed_time:.2f} секунд")

    # Вывод времени выполнения для каждого запроса
    for i, (uniprot_id, fetch_time) in enumerate(zip(uniprot_ids, fetch_times)):
        print(f"Запрос {i + 1} (ID: {uniprot_id}) выполнен за {fetch_time:.2f} секунд")

    # Построение графиков
    plt.figure(figsize=(12, 5))

    # График 1: Время выполнения каждого запроса
    plt.subplot(1, 2, 1)
    plt.plot(range(1, len(fetch_times) + 1), fetch_times, marker='o', linestyle='-', color='b')
    plt.title("Время выполнения каждого запроса")
    plt.xlabel("Номер запроса")
    plt.ylabel("Время (сек)")

    # График 2: Статистика успешных и неудачных запросов
    plt.subplot(1, 2, 2)
    plt.bar(["Успешные", "Неудачные"], [success, failure], color=["green", "red"])
    plt.title("Статистика запросов")
    plt.xlabel("Тип запроса")
    plt.ylabel("Количество")

    plt.tight_layout()
    plt.show()

# Запуск программы
await plot_results()
```

```python
import aiohttp
import asyncio
import time
import pandas as pd
import matplotlib.pyplot as plt
import requests  # Для синхронных запросов

# Асинхронная функция для получения данных в формате FASTA
async def data_async(session, uniprot_id):
    query_link = f"https://rest.uniprot.org/uniprotkb/search?query={uniprot_id}&format=fasta" # ссылка запроса
    start_time = time.time()  # замер времени начала запроса
    try:
        async with session.get(query_link) as response: # асинхронный HHTP-запрос - отправляет запрос на query_link, возвращает ответ от сервера
            # as response - присвоение результата выполнения запроса в переменную response

            if response.status == 200: # проверка, что запрос успешный 
                data = await response.text() # получение данных в формате FASTA
                # response.text - читает тело ответа в виде текста
                end_time = time.time()  # замер времени завершения запроса
                return data, end_time - start_time  # Возвращаем данные и время выполнения
            else:
                response.raise_for_status() # проверка, успешен ли запрос, если нет, то ошибка

    # обработка ошибок при запросе            
    except aiohttp.ClientError as e: # перехват всех ошибок, связанных с HTTP
        # as e - присвоение перехваченных ошибок переменной е
        end_time = time.time()  # замер времени завершения запроса (в случае ошибки)
        print(f"Произошла ошибка при запросе {uniprot_id}: {str(e)}")
        return None, end_time - start_time  # возвращаем None и время выполнения (в случае ошибки)

# Синхронная функция для получения данных в формате FASTA
def uniprot_data_sync(uniprot_id):
    query_link = f"https://rest.uniprot.org/uniprotkb/search?query={uniprot_id}&format=fasta"
    start_time = time.time()  # Замер времени начала запроса
    try:
        response = requests.get(query_link) # синхронный GET HTTP-запрос
        # ответ сохраняется в response
        if response.status_code == 200:

            data = response.text # тело ответа в виде текста
            end_time = time.time()  # Замер времени завершения запроса
            return data, end_time - start_time  # Возвращаем данные и время выполнения
        else:
            response.raise_for_status()

    # обработка ошибок при запросе        
    except requests.RequestException as e:
        end_time = time.time()  # Замер времени завершения запроса (в случае ошибки)
        print(f"Произошла ошибка при запросе {uniprot_id}: {str(e)}")
        return None, end_time - start_time  # Возвращаем None и время выполнения

# Чтение идентификаторов из файла Excel
def read_uniprot_ids(file_path):
    df = pd.read_excel(file_path)  # Чтение Excel-файла в DataFrame
    return df["Entry"].tolist()  # преобразование столбца Entry в список идентификаторов

# Асинхронная функция для выполнения запросов
async def run_requests_async(uniprot_ids):
    start_time = time.time()  # Общее время начала выполнения всех запросов
    async with aiohttp.ClientSession() as session: # создание асинхронной HTTP-сессии
        tasks = [data_async(session, uniprot_id) for uniprot_id in uniprot_ids] # для каждого идентификатора вызывается функция data_async(session, uniprot_id), которая возвращает задачу
        results = await asyncio.gather(*tasks) # параллельное выполнение всех задач из tasks
    end_time = time.time()  # Общее время завершения всех запросов

    # Разделяем результаты и время выполнения
    fetch_times = [result[1] for result in results]  # извлечение времени выполнения каждого запроса

    # Накопление времени выполнения
    cumulative_times = []
    total_time = 0
    for fetch_time in fetch_times:
        total_time += fetch_time
        cumulative_times.append(total_time)

    return cumulative_times

# Синхронная функция для выполнения запросов
def run_requests_sync(uniprot_ids):
    start_time = time.time()  # Общее время начала выполнения всех запросов
    results = [uniprot_data_sync(uniprot_id) for uniprot_id in uniprot_ids] # последовательное выполнение запросов
    end_time = time.time()  # Общее время завершения всех запросов

    # Разделяем результаты и время выполнения
    fetch_times = [result[1] for result in results]  # Время выполнения каждого запроса

    # Накопление времени выполнения
    cumulative_times = []
    total_time = 0
    for fetch_time in fetch_times:
        total_time += fetch_time
        cumulative_times.append(total_time)

    return cumulative_times

# Функция для построения графиков
def plot_results(async_times, sync_times):
    plt.figure(figsize=(12, 6))

    # График 1: Сравнение асинхронного и синхронного выполнения
    plt.plot(range(1, len(async_times) + 1), async_times, marker='o', linestyle='-', color='b', label="Асинхронные запросы")
    plt.plot(range(1, len(sync_times) + 1), sync_times, marker='o', linestyle='-', color='r', label="Синхронные запросы")
    plt.title("Сравнение асинхронного и синхронного выполнения")
    plt.xlabel("Количество запросов")
    plt.ylabel("Общее время (сек)")
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Основная функция
async def main():
    # Путь к вашему Excel-файлу
    input_file = "C:\\Users\\sasha\\OneDrive\\Рабочий стол\\infa\\uniprot.xlsx"
    
    # Чтение идентификаторов из файла
    uniprot_ids = read_uniprot_ids(input_file)

    # Выполнение асинхронных запросов
    async_times = await run_requests_async(uniprot_ids)

    # Выполнение синхронных запросов
    sync_times = run_requests_sync(uniprot_ids)

    # Построение графиков
    plot_results(async_times, sync_times)

# Запуск программы
await main()
```
